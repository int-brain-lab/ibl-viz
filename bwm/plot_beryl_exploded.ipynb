```python
import pandas as pd
import unityneuro.render as urn
import numpy as np
import imageio.v2 as imageio
```

```python
urn.setup(standalone=True)
```

```python
ontology = pd.read_csv('./ontology_structure_minimal.csv')
```

```python
ontology
```

```python
beryl_ids = np.unique(ontology['beryl_id'].values)
```

```python
urn.load_beryl_areas()
```

```python
def rgb_to_hex(rgb):
    return '#%02x%02x%02x' % rgb
rgb_to_hex((255, 255, 195))
```

```python
for beryl_id in beryl_ids:
    urn.set_area_material({str(beryl_id):'opaque-unlit'})
```

```python
id2color = {}
j = 0
for i, beryl_id in enumerate(beryl_ids):
    x = i
    y = 0
    if i > 255:
        j+=1
        x = 255
        y = j
    hex_color = rgb_to_hex((x,y,255))
    # print((beryl_id, hex_color))
    urn.set_area_color({str(beryl_id):hex_color})
    id2color[beryl_id] = (x,y,255)
```

```python
urn.set_camera_rotation(45,0,22.5)
```

```python

im = imageio.imread('exploded_color_test/exploded_left.png')
print(im.shape)
```

```python
im_flat = np.reshape(im,(im.shape[0]*im.shape[1],3))
```

```python
im_unique = np.unique(im_flat,axis=0)
```

```python
im_unique
```

```python
im_unique.shape
```

```python
found = 0
for val in im_unique:
    for id in id2color:
        if id2color[id][0]==val[0] and id2color[id][1]==val[1] and id2color[id][2]==val[2]:
            print((id2color[id],val))
            found+=1

print(found)
```
