```python
def get_color(val):
    return cm_data[np.min([int(val*255),255])]
```

```python
data = pd.read_csv('figure7_dataframe_filt.csv')
```

```python
avg_fr = data['avg_fr']
avg_fr = np.log10(avg_fr)
avg_fr = avg_fr - np.min(avg_fr)
avg_fr = avg_fr / np.max(avg_fr)
```

```python

bregma = atlas.xyz2ccf([0,0,0])

neuronCount = 0
neuronNames = []
neuronLocs = []
neuronColors = []
neuronSizes = []
neuronShapes = []

for (i, row) in data.iterrows():
    ccf = atlas.xyz2ccf([row.x + np.random.normal()/200000, row.y + np.random.normal()/200000, row.z + np.random.normal()/200000])
    
    if row.region=='PPC':

        neuronLocs.append([ccf[1], ccf[0], ccf[2]])
        neuronSizes.append([0.015, 0.015, 0.015])
        neuronShapes.append('sphere')
        
        # color
        color = get_color(avg_fr[i])
        neuronColors.append(color[:3])
```

```python
urchin.setup()
```

```python
urchin.ccf25.load()
```

```python
urchin.ccf25.VISa.set_visibility(True, urchin.utils.Side.LEFT)
urchin.ccf25.VISa.set_material('transparent-lit', urchin.utils.Side.LEFT)
urchin.ccf25.VISa.set_alpha(0.25, urchin.utils.Side.LEFT)
urchin.ccf25.VISam.set_visibility(True, urchin.utils.Side.LEFT)
urchin.ccf25.VISam.set_material('transparent-lit', urchin.utils.Side.LEFT)
urchin.ccf25.VISam.set_alpha(0.25, urchin.utils.Side.LEFT)
```

```python
urchin.camera.main.set_mode('perspective')
urchin.camera.main.set_rotation([22.5,22.5,225])
urchin.camera.main.set_zoom(8)
urchin.camera.main.set_pan(1.75,1)
```

```python
neurons = urchin.meshes.create(len(neuronLocs))
```

```python
urchin.meshes.set_positions(neurons, neuronLocs)
urchin.meshes.set_scales(neurons, neuronSizes)
urchin.meshes.set_colors(neurons, neuronColors)
```
